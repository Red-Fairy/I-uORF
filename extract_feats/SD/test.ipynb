```python
from ldm_extractor import LdmExtractor
import torch
import pickle
model = LdmExtractor().cuda()
model.eval()
model.training = False
data = {'img': torch.randn(4, 3, 256, 256).cuda(), 'text': 'hello world'}
out = model(data)
save = []
```

```python
data = {'img': torch.randn(4, 3, 256, 256).cuda(), 'text': 'hello world'}
model(data)
```

```python
for i in out:
    save.append(i.cpu().detach().numpy())
pickle.dump(save, open('test.pkl', 'wb'))
```

```python
out_load = pickle.load(open('test.pkl', 'rb'))
for i in out_load:
    print(i.shape)
```
